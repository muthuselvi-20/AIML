```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("shree1992/housedata")

print("Path to dataset files:", path)
```

```python
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.read_csv(path + "/data.csv")
df.head()
```

```python
#droping column
df.drop("date", axis=1, inplace=True)
```

```python
df.drop("statezip",axis = 1 , inplace=True)
```

```python
df.drop("street",axis = 1 , inplace=True)
```

```python
#unique values in column
df["country"].unique()
```

```python
df["country"] = df["country"].replace("USA",0).infer_objects(copy=False)
```

```python
df.drop("city", axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_error
```

```python
# independent variable
x = df[["bathrooms","bedrooms","sqft_living","sqft_lot","floors"]]
# dependent variable
y = df[["price"]]
```

```python
#splitting the data
x_train, x_test, y_train , y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
#train model
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
# test
y_predict = model.predict(x_test)
```

```python
print((r2_score(y_test, y_predict))*100)
print(mean_absolute_error(y_test, y_predict))
print(mean_squared_error(y_test, y_predict))
```
