1. STUDENT MARKS

```python
import numpy as np
```

```python
marks = np.array([[67,56,34,43,56],
                 [45,52,25,79,69],
                 [89,85,72,84,93]])
```

```python
# total no. of students
print("Students: ", np.size(marks,0))

# total no. of subjects
print("Subjects: ", np.size(marks,1))
```

```python
print(np.shape(marks))
print(marks.ndim)
```

```python
# First student's 2nd subject mark
print(marks[0,1])
```

```python
# lengthy
for i in range(np.size(marks,0)):
      average = np.mean(marks[i])
      print(f"Average of student {i+1} : ", average)
```

```python
#short
avg = np.mean(marks,1)
for i in range(len(avg)):
  print("Average of student", i+1,":",avg[i])
```

```python
minimum = np.min(marks,0)
print(minimum)

#extra mark for low marks students
minimum += 10
print("After + 10")
print(minimum)
```
